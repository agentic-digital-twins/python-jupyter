# 🧠 ISO-NE API Aggregator Snippet for Jupyter Notebook

##Here’s a Python code snippet that creates a reusable and notebook-friendly data aggregator for the ISO-NE energy and market data API (v1.1). 
It handles:

Basic Auth

Endpoint access

Error handling

Pandas DataFrame conversion (ideal for Jupyter)

{
MonthlyNetworkLoads: {
MonthlyNetworkLoad: [
{
BeginDate: "2025-05-01T00:00:00.000-04:00",
NetworkLoadId: 2,
NetworkLoadName: "Braintree",
CustomerId: 5,
CustomerName: "Braintree Electric Light Depar",
DunsNumber: "17-057-1897",
DunsName: "Braintree Electric Light Dept.",
ReliabilityRegionId: 9006,
ReliabilityRegionName: ".R.SEMASS",
LocalNetwork: "BE   ",
NetworkLoadValueMw: 37.143,
PoolNetworkLoadValueMw: 14277.648
},
]}}


```python
import requests
import pandas as pd
from requests.auth import HTTPBasicAuth
import getpass 

class ISONEDataAggregator:
    def __init__(self, username: str, password: str):
        self.base_url = "https://webservices.iso-ne.com/api/v1.1"
        self.auth = HTTPBasicAuth(username, password)

    def fetch_json(self, endpoint: str, params: dict = None) -> dict:
        """Fetch data from a specific ISO-NE API endpoint with .json suffix."""
        url = f"{self.base_url}/{endpoint}.json"
        response = requests.get(url, auth=self.auth, params=params)
        if response.status_code != 200:
            raise Exception(f"Request failed: {response.status_code} - {response.text}")
        return response.json()

    def fetch_as_dataframe(self, endpoint: str, params: dict = None) -> pd.DataFrame:
        """Return a pandas DataFrame from an endpoint that returns a JSON array."""
        data = self.fetch_json(endpoint, params)
        
        # Flatten top-level array fields
        if isinstance(data, list):
            print('list')
            return pd.DataFrame(data)
        elif isinstance(data, dict):
            print('dict')
            # Try to find the main array-like key
            for key, value in data.items():
                if isinstance(value, list):
                    print('dict-list')
                    return pd.DataFrame(value)
            print('single record fallback')
            # Extract the list
            network_loads = data["MonthlyNetworkLoads"]["MonthlyNetworkLoad"]
            
            # Create DataFrame
            return pd.DataFrame(network_loads)
            #return pd.json_normalize(data)
            #return pd.DataFrame([data])  # single record fallback
        else:
            raise ValueError("Unexpected JSON structure returned by API")

# Example usage inside Jupyter
username = input("your_iso_ne_username")
password = getpass.getpass("your_iso_ne_password")

iso = ISONEDataAggregator(username, password)

# Example: Get current network load
df_load = iso.fetch_as_dataframe("networkload/current")
#df_load.head()
df_load

```
